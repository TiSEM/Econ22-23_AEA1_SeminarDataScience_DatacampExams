# Paper (0.5 points)

**The paper used in this assignment is Deaton, A. (1991). Saving and Liquidity Constraints. _Econometrica_, 59(5), 1221–1248. https://doi.org/10.2307/2938366**

This paper focuses on building a dynamic programming model to investigate consumer consumption maximization behavior under the constraints of asset accmulation and exogenous labor income. In the author's framework, a consumer maximizes lifetime utility based solely on how much to consume in each period. The consumer becomes happier when she consumes more, but she cannot consume as much as she wants because there are some restrictions. These restrctions are first, the amount of consumption in the current period influences the amount of assets avaliable for the next period, and the amount of asset is always required to be non-negative so the consumer is free from any debt. Second, the consumer receives stochastic labor income in each period and she cannot influence how much she receives. Finally, a decision has to be made by the consumer before the game starts. She knows how patient she is in terms of how much she likes to consume now an later, the growth of asset accmulation and how the labor income is distributed such that she can estimate approximately how much she receives. She can also see all combinations of how much to consume for each period, and she selects the best plan that will make her the happinest life. The author believes these simple settings approximate consumer behavior in reality, and this paper proves to be successful in explaining certain economic phenomena.

A dynamic programming model quantifies the situation described above. Generally, a dynamic programming model specifies a maximization problem, which is typically the sum of utility from the beginning to the end or infinity, and some constraints that binds each two adjecent peirods. The solution of this model is a path, consistuted by how much to consume in each period that subsequetly leads to the highest utility and how much assets she has in each period. Researchers are interested to see how variations in some parameters will influence this path.

However, the research question in this paper is quantitative. The author investigates how different statistical forms of income influence consumption, and assumes different data generating processes for the income, such as stationary distribution and random walk. The results involve some numerical estimations of statistical moments such as autocorrelations and standard deviations, which are typical interests for macroeconomists in central banks and ministries of finance, but not for general audiences of this assignment. **Briefly speaking, this paper shows the behavior of saving and asset accumulation is quite sensitive to what consumers believe about the stochastic process generating their incomes.**

# Main assumptions (0.5 points)

This section will briefly introduces deterministic dynamic programming in discrete time and a solution method. This introduction relys on *Further mathematics for economic analysis.* Consider a system that changes at discrete times $t = 0, 1, ..., \infty$. Assume that the initial state $x_0$ is historically given, and that from then on the system evolves through time under the influence of a sequence of controls $u_t$, which can be chosen freely from a given set $\mathrm{U}$. The controls influence the evolution of the system through a difference equation

\begin{equation}
x_{t+1} = g(t, x_t, u_t), \quad x_0 \ \text{given,} \quad u_t \in \mathrm{U}
\end{equation}

Assume that there is a function $f(t,x,u)$ of three variables such that the utility associated with a given path is represented by the sum

\begin{equation}
\sum^{\infty}_{t=0} f(t,x,u)
\end{equation}

The sum is called the objetive function. The discrete time optimization problem can be briefly formulated as

\begin{equation}
\max \sum^{\infty}_{t=0} f(t,x,u) \quad \text{subject to} \quad x_{t+1} = g(t, x_t, u_t), \quad x_0 \ \text{given,} \quad u_t \in \mathrm{U}
\end{equation}

**`Main assumption 1: $f(t,x,u)$ is concave.`** 

**`Main assumption 2: $f(t,x,u)$ is compact, and this means`**

\begin{equation}
M_1 \leq f(t,x,u) \leq M_2 \quad \text{for all} \ (x,u) \ \text{with} \ u \in \mathrm{U}
\end{equation}

**where $M_1$ and $M_2$ are given numbers. Under these two conditions, this maximization problem has a solution.**

A solution method is to apply Bellman equations. Define the value function for the problem at time $s$ by

\begin{equation}
V_s(x) = \max_{u \in \mathrm{U}} \sum^{\infty}_{t=s} f(t,x,u)
\end{equation}

And the value function satisfies the equation

\begin{equation}
V(x) = \max_{u \in \mathrm{U}} \big[f(x,u)+ \beta V(g(x,u))\big] \tag{the Bellman equation}
\end{equation}

**To put it simply, there exist a function that captures the maximum value of the sum of the utilities from the subsequent period till the end, and we can write down the maximization problem as to maximize the sum of current utility and that function. Then we write down this maximization problem for all periods, and we get a system of equations.**

Normally, the functional form of value functions are unknown. A numerical method is to assume an initial value $V_0$, and interate over all periods.

**`The specific assumptions in the model from the paper are`**

**`In my analysis, I assume`**

# Model in the paper (1.0 points)

The consumer maximizes the utility function

\begin{equation}
u = \mathbf{E}\big\{\sum^{\infty}_{\tau = r}(1+\delta)^{t-\tau} v(c_\tau)\big\}
\end{equation}

where $\delta > 0$ is the rate of time preference, and $v(c_t)$ is the instantneous utility function, assumed to be increasing, strictly concave, and differentiable. The evolution of assets is given by

\begin{equation}
A_{t+1} = (1+r)(A_t + y_t - c_t)
\end{equation}

where $y_t$ is labor income, $A_t$ is real assets, and $r$ is the real interest rate. Labor is inelastically supplied, and $y_t$ is a stationary random variable with support $[y_0, y_1]$, with $y_0 > 0$ and $y_0 \leq y_1 \leq \infty$; income cannot fall below the positive floor $y_0$. The borrowing restriction is

\begin{equation}
A_t \geq 0.
\end{equation}

Moreover, the consumption is non-negative, $c_t \geq 0$.

This is a stochastic dynamic programming. Despite that, the general approach is the same as previously introduced. The Bellman equation for this problem, as indicated in the author's paper, is

\begin{align*}
V_t(x_t) & = \max_{c_t \in [0,x_t]} \big[ u(c_t) + \beta \int V_{t+1}(x_{t+1} p(d x_{t+1}|x_t,c_t)\big]\\
 & = \max_{c_t \in [0,x_t]} \big[ u(c_t) + \beta \int V_{t+1}((1+r)(A_t + y_t - c_t)g(y_{t+1})d y_{t+1}\big]
\end{align*}

where $g(y)$ is the density function of $y$, $ln(y_t) \sim N(\mu, \sigma^2)$. The solution is a sequence of pairs $\{A_t, c_t\}, t = 1, ..., \infty$.

**To put it simply, previously we have the a problem that needs to maximize the sum of the current utility and a function that describes the maximal value of the sum of all future utilities. Now we have a similar problem that needs to maximize the sum of the current utility and a function that describes the expected value of the maximal value of the sum of all future utilities, hence we need to integrate over the distribution of the random variable, in this case, stochastic labor income.**

**The small extension I introduce is different parametric values for the real interest rate. In the author's investigation, the interest rate is fixed, so there is no research into the response of consumption to interest rate. In this assignment, I vary the interest rate, and explore how consumption responses.**

# Python code (6.0 points)

*Studying these online materials, I learn that it is more convenient to create a class for a model. A class in like a package. It can have a custom name and it bundles multiple functions together. The dot method can also be specified in a class. So the usage of a class is not different from a package. If it is designed well, the complex model can be process just by calling this class. The class is both used in QuantEcon and Schjerning's codes. Here I apply the main framework by Schjerning.*

```python
import numpy as np
import scipy.stats
import scipy.interpolate
from time import process_time
import matplotlib.pyplot as plt

class DynamicProgram():
    '''First, the utility function. And it takes log form, same as standard economic analysis.'''
    def u(self,c):
        return np.log(c)
    
    '''This is an instantiation operation. It can creates methods specified in this class,
    and values can be directly obtained when methods are called. Default values can also be
    specified in the brackets. Here there are no default values.'''
    def __init__(self,β, r, μ, σ, xbar, n_x, n_c, n_y):
        # Structral parameters
        self.β  = 1/(1+β)
        self.R  = 1+r
        self.μ  = μ
        self.σ  = σ
        
        '''The following codes are preparations for Gauss-Legendre quadrature method, which is a method to compute the expected value of a distribution. 
        Schjerning uses this method, so I directly apply it. It is essentially a method to approximate integration
        (dividing the domain into many small intervals, selecting a point, x, in each interval and getting f(x). Then the integration is determined by
        an algorithem to calculate the sum of f(x) times length of intervals).'''
        
        # Number of grid points. The larger the number, the smaller the interval.
        self.xbar = xbar  # Upper and lower bound on asset
        self.n_x = n_x    # Number of grid points for asset
        self.n_c = n_c    # Number of grid points for consumption
        self.n_y = n_y    # Number of quadrature points for income
        
        '''Here are the creation of grids for integration approximation. Same as the Newton method for
        root finding in the class, we don't need to understand it. So I directly use Schjerning's codes and comments.'''
        # quadrture grids for y (adjusted weights and nodes)
        q, w = np.polynomial.legendre.leggauss(n_y) # legendre quadrture nodes and weights on [-1,1]
        Ginv = lambda z:  np.exp(scipy.stats.norm.ppf(z, loc=self.μ, scale =self.σ))  # inverse cdf of log-normal
        self.weights=w/2; # change of varibale to adjust to [0,1] interval
        self.y = Ginv((q+1)/2) # use change of varibale to adjust nodes to [0,1] interval and use Ginv to obntain y
        
        # grids for x and c (adjusted to satisfy constraints)
        self.xbar[0] = np.maximum(np.finfo(float).eps, self.xbar[0]);  # truncate lower bound at smallest positive float number
        self.x = np.linspace(self.xbar[0],self.xbar[1],n_x).reshape((n_x,1)) # grid for state space (n_w x1 array)
        self.c = np.empty((n_x,n_c)) # initilize grid for choices for each staty point 
        for i in range(n_x): 
            # make grid between that satisfies the constraint, c\in[0,x[i]] 
            self.c[i,:] = np.linspace(self.xbar[0],self.x[i],n_c).reshape((1,n_c)) 
            
    '''Here codes for Bellman equation are also directly learnt from Schjerning.'''
    def bellman(self,V0):
        interp = scipy.interpolate.interp1d(self.x[:,0],V0,  bounds_error=False,fill_value='extrapolate')
        EV=0 
        for i, y_i in enumerate(self.y): # compte expectation wrt to shocks to future state
            x1=self.R*(self.x-self.c) + y_i # next period x, conditional of w and c, shock, nW x nC
            EV+=self.weights[i]*interp(x1);
        matV1 = self.u(self.c) + self.β * EV        
        i_max= np.argmax(matV1,axis=1); # (column) idex of optimal choices
        V1 = matV1[np.arange(self.n_x),i_max]  
        c1 = self.c[np.arange(self.n_x),i_max] 
        return V1, c1

'''Dynamic Programming solver.'''
class dpsolver():
    def vfi_T(model, T=100, callback=None):
        '''Solves the model using backward induction (VFI with maxiter =T'''
        tic = process_time() # Start the stopwatch / counter
        V=np.zeros((model.n_x, T+1)) # on first iteration assume consuming everything
        c=np.zeros((model.n_x, T+1)) # on first iteration assume consuming everything
        for t in range(T-1, 0, -1):
            V[:,t-1],c[:,t-1]=model.bellman(V[:,t])
            if callback: callback(t,model.x,V, c) # callback for making plots and plotting iterations
        else:  # when i went up to maxiter
            toc = process_time() # Stop the stopwatch / counter
            print(model, 'solved by backward induction using',round(toc-tic, 5), 'seconds')
        return V,c
    
    def vfi(model, maxiter=100, tol=1e-6,callback=None):
        '''Solves the model using VFI (successive approximations)'''
        tic = process_time() # Start the stopwatch / counter
        V0=np.zeros(model.n_x) # on first iteration assume consuming everything
        for iter in range(maxiter):
            V1,c1=model.bellman(V0)
            if callback: callback(iter,model.x,V1, c1, V0) # callback for making plots
            if np.max(abs(V1-V0)) < tol:
                toc = process_time() # Stop the stopwatch / counter
                print(model, 'solved by VFI in', iter, 'iterations using',round(toc-tic, 5), 'seconds')
                break
            V0=V1
        else:  # when i went up to maxiter
            print('No convergence: maximum number of iterations achieved!')
        return V1,c1
    
    def iterinfo(iter,model,V1,c=None, V0=0):
        print('iter=', iter, '||V1-V0||', np.max(abs(V1-V0)))
        
'''This is a simulator'''
def simulate(model,Tsim=99, T=100, A0=0):
    V, c = dpsolver.vfi_T(model, T);  
    x=model.x; 
    c_t = np.empty(Tsim)
    x_t = np.empty(Tsim)
    y_t = np.exp(np.random.normal(loc=model.μ, scale=model.σ, size=Tsim))
    x_t[0]=A0+y_t[0];
    for t in range(Tsim):
        c_interp = scipy.interpolate.interp1d(x[:,0],c[:,t],  bounds_error=False,fill_value='extrapolate')
        c_t[t] = c_interp(x_t[t])
        if t<Tsim-1:
            x_t[t+1] = model.R*(x_t[t]-c_t[t]) + y_t[t+1]
    return y_t, x_t, c_t
```

Note 1

Gauss-Legendre Quadrature
\begin{equation}
\int^{1}_{-1}f(x)dx = \sum^{n}_{i=1}w_i f(x_i) + \frac{2^{2n+1}(n!)^4}{(2n+1)!(2n)!}\frac{f^{(2n)}(\epsilon)}{(2n)!}
\end{equation}
for some $\epsilon \in [-1,1]$, or $[a,b]$.

Gauss-Legendre for expectations functions of random variables
\begin{equation}
\mathrm{E}[f(y)] = \frac{1}{2}\int^1_0 f\big(\frac{G^{-1}(x)+1}{2}\big)dx \approx \sum^{n}_{i=1}\frac{w_i}{2}\big(\frac{G^{-1}(x_i)+1}{2}\big)
\end{equation}
